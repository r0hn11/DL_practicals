TENSORFLOW

Import libraries

```python
import tensorflow as tf
import numpy as np
```

```python
a = tf.constant(15)
b = tf.constant(20)
print(a + b)
```

Input

```python
x = np.random.rand(100).astype(np.float32)
print(x)
```

Output

```python
y = x * 0.2 + 0.2
```

Weight

```python
w = tf.Variable(tf.random.normal([1]))
```

Bias

```python
b = tf.Variable(tf.zeros([1]))
```

Create a function for MSE - Mean Squared Error

```python
def mse_loss():
  ypred = w * x + b
  loss = tf.reduce_mean(tf.square(ypred - y))
  return loss
```

Optimizer

```python
optimizer = tf.keras.optimizers.Adam()
```

Iterations

```python
for step in range(5000):
  optimizer.minimize(mse_loss, var_list=[w, b])
  if step % 500 == 0:
    print(step, w.numpy(), b.numpy())
```

KERAS

Import libraries

```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
dataset = loadtxt('pima-indians-diabetes.csv', delimiter = ',')
dataset
```

```python
x = dataset[:, 0:8]
print(type(x))
print(x.shape)
```

```python
y = dataset[:, 8]
print(y)
```

```python
model = Sequential()
```

```python
model.add(Dense(12, input_shape=(8,), activation='relu'))
```

```python
model.add(Dense(8, activation = 'relu'))
```

```python
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.fit(x, y, epochs = 150, batch_size = 10)
```

```python
model.evaluate(x, y)
```

THEANO

Import libraries

```python
import theano
from theano import *
import theano.tensor as T
import numpy as np
import pandas as pd
from theano import function
```

Scalar variable

```python
v1 = T.dscalar()
v2 = T.scalar()
```

Subtraction

```python
sres = v1 - v2
```

Add

```python
ares = v1 + v2
```

Convert the results into functions

```python
calcsres = theano.function([v1, v2], sres)
calcares = theano.function([v1, v2], ares)
calcares(12, 23)
calcsres(13, 12)
x = T.dmatrix('x')
y = T.dmatrix('y')
```

Addition

```python
z = x + y
func = function([x, y], z)
m1 = [[1,2], [3,4]]
m2 = [[4,5], [6,7]]
func(m1, m2)
```

PYTORCH

Import libraries

```python
import torch
import numpy as np
```

Using data

```python
data = [[1,2], [3,4]]
x_data = torch.tensor(data)
print(type(x_data))
```

Using numpy arrays

```python
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
print(x_np)
print(type(x_np))
```

Using another tensor

```python
x_ones = torch.ones_like(x_data)
print("One Tensor:\n", x_ones)
x_rand = torch.rand_like(x_data, dtype = torch.float)
print(x_rand)
```

More ways to create tensor

```python
shape = (2, 3)
random_tensor = torch.rand(shape)
print(random_tensor)
print(type(random_tensor))
ones_tensor = torch.ones(shape)
print(ones_tensor)
print(type(ones_tensor))
zeros_tensor = torch.zeros(shape)
print(zeros_tensor)
print(type(zeros_tensor))
tensor = torch.rand(3, 4)
print(tensor)
tensor.shape
tensor.dtype
tensor.device
```

Tensor Operations

```python
if torch.cuda.is_available():
  tensor = tensor.to('cuda')
  print("Device tensor is stored in ", tensor.device)
```

Indexing, slicing

```python
tensor = torch.ones(4, 4)
print("Tensor:\n", tensor)
tensor1 = torch.zeros(4, 4)
print("Tensor1:\n", tensor1)
tensor2 = torch.cat([tensor, tensor1])
print("Tensor2:\n", tensor2)
```

Multiply Operation

```python
tensor.mul(tensor1)
tensor * tensor1
tensor.T
```

Inplace change the original tensor

```python
tensor.add_(3)
print(tensor)
```

From tensor to numpy

```python
t = torch.ones(5)
print(t)
n = t.numpy()
print(n)
print(type(n))
```
